```python
import os
import re
import string
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
```

```python
#preprocessing and cleaning
```

```python
#removing stopwords     
def text_process(content):
    print('processing')
    return [word for word in content.split() if word.lower() not in stopwords.words('english')]
```

```python
#fuction for data loading
```

```python
def loadData():
    data1=pd.read_csv('1.csv', sep='*')
    data2=pd.read_csv('2.csv', sep='*')
    data3=pd.read_csv('3.csv', sep='*')
    data4=pd.read_csv('4.csv', sep='*')
    data5=pd.read_csv('5.csv', sep='*')
    list=[data1,data2,data3,data4,data5]
    data=pd.concat(list)
    return data
```

```python
#function for describe
```

```python
def showDetails():
    result=loadData()
    result=result.sample(frac=1)
    print(result.head())
    print(result.groupby('class').describe())
```

```python
#loading data
```

```python
result=loadData()
print(result['content'])
```

```python
#countvectorization i.e. building a dictionary dict={ word: count}
```

```python
#importing countvectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#building countvectorizer
cv= CountVectorizer(analyzer=text_process).fit(result['content'])
```

```python
# Print total number of vocab words
print(len(cv.vocabulary_))
cv.get_feature_names()[900]
```

```python
#i.e. we have 38965 words in vocabulary
```

```python
#this is first article our dataset
article = result['content'].iloc[0]
print(article)
```

```python
#words in article and its count in countvectorizer
bow = cv.transform([article])
print(bow)
print(bow.shape)
```

```python
#4074th and 9571th features
print(cv.get_feature_names()[4073])
print(cv.get_feature_names()[9570])
```

```python
article_bow=cv.transform(result['content'])
```

```python
print('Shape of Sparse Matrix: ', article_bow.shape)
print('Amount of Non-Zero occurences: ', article_bow.nnz)
```

```python
sparsity = (100.0 * article_bow.nnz / (article_bow.shape[0] * article_bow.shape[1]))
print('sparsity: {}'.format((sparsity)))
```

```python
#IDF(t) = log_e(Total number of documents / Number of documents with term t in it)
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer().fit(article_bow)
```

```python
articles_tfidf = tfidf_transformer.transform(article_bow)
print(articles_tfidf.shape)
```

```python
#Training a model using MultiNomialNB
from sklearn.naive_bayes import MultinomialNB
classifier_model = MultinomialNB().fit(articles_tfidf, result['class'])
```

```python
predictions = classifier_model.predict(articles_tfidf)
print(predictions)
```

```python
from sklearn.metrics import classification_report
print (classification_report(result['class'], predictions))
```

```python
#splitting data into train and test data
from sklearn.model_selection import train_test_split

article_train, article_test, class_train, class_test = \
train_test_split(result['content'], result['class'], test_size=0.2)

print(len(article_train), len(article_test), len(class_train) + len(class_test))
```

```python
#creating pipeline
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('bow', CountVectorizer()),  # strings to token integer counts
    ('tfidf', TfidfTransformer()),  # integer counts to weighted TF-IDF scores
    ('classifier', MultinomialNB()),  # train on TF-IDF vectors w/ Naive Bayes classifier
])
```

```python
#classification report
predictions = pipeline.predict(article_test)
print(classification_report(predictions,class_test))
predictions
```

```python
pipeline.fit(article_train,class_train)
from sklearn.externals import joblib
joblib.dump(pipeline, 'pipe.pkl') 
```

```python
#this function predicts the class of given article
import string
import pandas as pd
from nltk.corpus import stopwords

def get_category(pred):
    dict = { 1: 'Business', 2: 'Entertainment', 3: 'Politics',4: 'Sports', 5: 'Technology'}
    return dict[pred[0]]

def text_process(content):
    punct_rem = [char for char in content if char not in string.punctuation]
    punct_rem = ''.join(punct_rem)
    return [word for word in punct_rem.split() if word.lower() not in stopwords.words('english')]

from sklearn.externals import joblib
pip= joblib.load('pipe.pkl') 

with open('README.TXT', 'r') as content_file:
    content = content_file.read()
    
content=content.replace('\n', ' ').replace('\r', '')
content=' '.join(text_process(content))

with open('README.TXT', 'w') as content_file:
    content_file.write(content)
    
test=pd.read_csv("README.TXT",sep='*')
pip=pip.predict(test)

print(get_category(pip))
```
